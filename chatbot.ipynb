```python
# Step 1: Import necessary libraries and load the PDF file
from langchain_community.document_loaders import PyPDFLoader

# Update the file path for Colab (adjust if using Google Drive)
pdf_path = "/content/drive/MyDrive/attention.pdf"  # Replace with the actual file path
loader = PyPDFLoader(pdf_path)
```

```python
# Load the PDF document
docs = loader.load()
print("Loaded documents:", docs)
```

```python
# Step 2: Split the document into chunks
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Configure the text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)

# Split the loaded document into chunks
chunk_documents = text_splitter.split_documents(docs)
print(f"Number of chunks created: {len(chunk_documents)}")
```

```python
from huggingface_hub import login

# Paste your token here
login(token="Enter your hugging face access token here")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
# from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
# Use SentenceTransformers model for embeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Create FAISS Vector Store
db = FAISS.from_documents(chunk_documents[:30], embeddings)
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load Hugging Face model (e.g., LLaMA 2 or Falcon)
model_name = "gpt2"  # or another open-source model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
llm_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, max_length=1024)

# Create HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=llm_pipeline)
```

```python
# Step 6: Design a chat prompt template
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
Answer the following question based only on the provided context.
Think step by step before providing a detailed answer.
<context>
{context}
</context>
Question: {input}""")
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

# Document Chain
document_chain = create_stuff_documents_chain(llm, prompt)

# Retrieval Chain
retriever = db.as_retriever()
from langchain.chains import create_retrieval_chain
retrieval_chain = create_retrieval_chain(retriever, document_chain)

# Query
response = retrieval_chain.invoke({"input": "what are Encoder and Decoder Stacks?"})
```
